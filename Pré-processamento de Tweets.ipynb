### **Criando stop list e substituindo emotes e caracteres especiais ;)**

```python
stop_list_nltk = nltk.corpus.stopwords.words('english')
stop_list_nltk.append('rt')
```

```python
def sub(sentence):
  sentence = re.sub(r'http(.*?)\s', '', sentence)
  sentence = re.sub(r':\)', 'happy', sentence)
  sentence = re.sub(r':D', 'happy', sentence)
  sentence = re.sub(r'XD', 'happy', sentence)
  sentence = re.sub(r';\)', 'happy', sentence)
  sentence = re.sub(r'>_<', 'happy', sentence)
  sentence = re.sub(r':\(', 'sad', sentence)
  sentence = re.sub(r';\(', 'sad', sentence)
  sentence = re.sub(r';-;', 'crying', sentence)
  sentence = re.sub(r';_;', 'crying', sentence)
  sentence = re.sub(r'\'-\'', 'annoyed', sentence)
  sentence = re.sub(r':/', 'annoyed', sentence)
  sentence = re.sub(r'\\u2019', '\'', sentence)
  sentence = re.sub(r'u002c', ',', sentence)

  return sentence
```

Criando dicionario

```python
dicio = dict()
```

### **Definindo função de tokenização que separa palavrs, remove emotes, stop words e lematiza**

```python
def tokeniza(sentence):
  sentence = sentence.lower()
  tokens = re.split(r'[-&|\s!$@#,.:;\\\"_?\)\(\]\[/]+',sentence)

  novel_tokens=[]
  for t in tokens:
    if t not in stop_list_nltk and t != '':
      novel_tokens.append(t)
      if t in dicio:
        dicio[t] = dicio[t]+1
      else:
        dicio[t] = 1

  return [lemmatizer.lemmatize(t) for t in novel_tokens]

```

```python
sentences = dataframe['text']
sentences[0]
```

```python
final_tweets = []
```

### **Aplicando tokenização nas sentenças**

```python
for sentence in sentences:
  sentence = sentence + ' '
  sentence = sub(sentence)
  final_tweets.append(tokeniza(sentence))
```
